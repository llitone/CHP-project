```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
data = pd.read_csv("db.csv")
```

Shishlyannikova_t1.jpg

```python
df = data.corr()
itog = []

for i in df:
  for j in df:
    if df[i].corr(df[j]) >= 0.7 and df[i].corr(df[j]) != 1.0:
      itog.append(f"{i} - {j}: {df[i].corr(df[j])}")

for i in list(set(itog)):
  print(i)
  print("\n")
```
